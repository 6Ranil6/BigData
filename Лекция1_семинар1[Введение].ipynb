## Лекция 1

<u>Проблемы и решения, которые возникают, когда данных очень много:</u>

1) Объем -> Увеличить вычислительные ресурсы
2) Проверка достоверности данных / Валидация данных -> нет единого подхода, некоторые даже руками начинают проверять, некоторые пытаются автоматизировать это все
3) Структурирование данных -> СУБД, key-value
4) Предварительная обработка -> распараллеливание (<b>это краеугольный камен, которым будем заниматься на этом занятии</b>)



<u>Примечание:</u>
- Можно дать пользователю возможность подстраивать модель под данные...
- Или же можно сделать так, чтобы модель сама итерационно проходила по всем 4-м пунктам описанных выше...



Пролема машинного обучения:
У нас нет генеральной совокупности, т.е. нет всех данных, которые могут появляться в жизни и если это появится, то модель уже ненужна будет.

<u>Как выглядят данные:</u>

Все собирают много данных, а для модели нужно малое кол-во данных, т.к. с росстом данных вероятность ошибки данных не увеличивается, т.е. главная задача выделить правильный кусок данных из всех собранных данных и обучить модель на этих данных.

<u>Примечание для картины сверху:</u>

Раньше данными работали, как слева показано, а теперь решили, что нужно обрабатывать данные, там же где они храняться. И один из плюсов - это когда скорость интернета благодаря этому ускоряется, т.к. трафик становитсья быстре/е


<u>Что нужно учитывать?</u>

Тот кто все создавал был писеммистом и они сказали, пусть у нас будет:

- Сеть не обязательно надежна
- Существуют задрежки
- Пропускная способность ограничена
- Топология может изменяться
- Может быть несколько администраторов
- Существуют транспортные затраты
- Сеть может быть гетерогеенной



И они поняли, что должен быть master, в котором данные не обрабатываются, а он только регулируют, где и как храняться данные
