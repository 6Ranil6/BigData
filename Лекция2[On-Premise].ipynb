# Занятие 2. Распределенная система

<u>Как происходит вычисление над большими данными?</u>

1) Отложенная обработка. (batch-овая) Накапливая пачку документов и только потом обрабатываем
2) Псевдореального времени.  Для обработки данных моментально, но по техническим обстаятельствам мы сразу обрабатывать данные не можем. Алгоритмы псевдореального времени помогает скрывать нужное время для обработки, а для пользователя это выглядит так, что мы сразу все обрабатываем. Любую отложенную обработку можно свести к псевдореального времени просто уменьшить размер батчей
3) Реального времени. Все к этому стремяться, но очень редко достигаются.

<u>Примечание:</u>
- Пункты 1 и 2 -  это *пакетная обработка / пакетные вычисления*
- Пункт 3 - это *потоковая обработка*

На занятии мы будем рассматривать в основном 1 и 2.

<u>Для пакетной обработки используется:</u>

- SQL (PostgreSQL, MySQL и т.д.)
- NoSQL (HBase, Casandra)
